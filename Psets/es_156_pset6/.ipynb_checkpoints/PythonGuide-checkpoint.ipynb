# Basic Sampling and Plotting with Pytplot

**Further documentation on using matplotlib/pyplot:**
- Basic plotting tutorials and examples: https://matplotlib.org/tutorials/index.html


```python
# frequency in Hz of the signal
nu = 25.0

#samples per seconds and sample rate
sampling = 1024
dt = 1.0 / sampling

# total number of samples
nsample = 500

# Plot a basic sine wave in the time domain
xt = np.linspace(0.0, nsample*dt,nsample)
yt = np.sin(nu * 2.0*np.pi*xt)
plt.plot(xt, (yt[0:nsample]))
plt.xlabel("s")
plt.grid()
plt.show()
```

# FFT and Plotting in the Frequency Domain

**Further documentation on using the fft:**
- Using numpy's fft function: https://docs.scipy.org/doc/scipy/reference/fftpack.html
- Many other useful fft related functions: https://docs.scipy.org/doc/numpy/reference/routines.fft.html

```python
# Plot the signal's FFT in the frequency Domain (same signal as above)
yf = fft(yt)
xf = np.linspace(0.0, 1.0/(2.0*dt), nsample/2)
plt.plot(xf, 2.0/nsample * np.abs(yf[0:nsample/2]))
plt.xlabel("Hz")
plt.grid()
plt.show()
```

# Spectrograms and Subplots

**Further documentation on the mlab.specgram function and subplotting:**
- Spectrogram: https://matplotlib.org/api/_as_gen/matplotlib.axes.Axes.specgram.html#matplotlib.axes.Axes.specgram
- Subplots with pyplot: https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplots.html
- More examples of using subplots: 
    - https://matplotlib.org/gallery/lines_bars_and_markers/categorical_variables.html#sphx-glr-gallery-lines-bars
    - https://matplotlib.org/gallery/lines_bars_and_markers/csd_demo.html#sphx-glr-gallery-lines-bars-and-markers-csd-demo-py
- Another spectrogram plotting example: https://pythontic.com/visualization/signals/spectrogram

```python
import matplotlib.pyplot as plt
import numpy as np

'''
To demonstrate the functionality of the spectrogram function,
we start by making a random signal. This signal and the steps 
to produce it are not relevent, so feel free to ignore this
initial part.
'''
# Fixing random state for reproducibility
np.random.seed(19680801)

dt = 0.0005
t = np.arange(0.0, 20.0, dt)
s1 = np.sin(2 * np.pi * 100 * t)
s2 = 2 * np.sin(2 * np.pi * 400 * t)

# create a transient "chirp"
mask = np.where(np.logical_and(t > 10, t < 12), 1.0, 0.0)
s2 = s2 * mask

# add some noise into the mix
nse = 0.01 * np.random.random(size=len(t))
"""
End of making our signal 
(More important part below)
"""


x = s1 + s2 + nse  # the signal

'''
The specgram function takes many parameters.
Many of these are useful, but also optional. We 
define a few that we will use below.
'''
NFFT = 1024  # the length of the windowing segments
Fs = int(1.0 / dt)  # the sampling frequency


# Figures and subplots allow us to display multiple plots at a time
fig, (ax1, ax2) = plt.subplots(nrows=2)
fig.subplots_adjust(hspace=0.5)

# Plot the signal in our first plot 
ax1.plot(t, x)
ax1.set_title("Original Signal")

# Plot the spectrogram in our second plot
Pxx, freqs, bins, im = ax2.specgram(x, NFFT=NFFT, Fs=Fs, noverlap=900)
ax2.set_title("Spectrogram")

# The `specgram` method returns 4 objects. They are:
# - Pxx: the periodogram
# - freqs: the frequency vector
# - bins: the centers of the time bins
# - im: the matplotlib.image.AxesImage instance representing the data in the plot
plt.show()
```
