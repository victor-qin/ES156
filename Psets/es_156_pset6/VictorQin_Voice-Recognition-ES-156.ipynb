## Data cleanup
First we start by importing a few packages we will use.

```python
import numpy as np
import numpy.random as random
from numpy.fft import fft
from scipy.io import wavfile
import matplotlib.pyplot as plt
import seaborn as sns
import os
%matplotlib inline
sns.set()
sns.set(font_scale=1.5)
```

Next we load the data. By changing the digits in the list `digit`, you can import different spokent digits from the dataset.

The recordings are all sampled at 8KHz. The recordings, however, have different lengths due to different recording durations. The maximum vector length will determine the length of the transform (shorter vectors will be padded with zeros). This is given by the variable `N`.

The dictionary `samples` will store the recorded samples. So `samples[d]` returns a list of vectors that contains recordings for digit `d`. 

```python
data_dir = '../es_156_pset6/data/free-spoken-digit-dataset-master/recordings/'

# determine digits of interest (0 to 9)
digits = [1,2,3,4,5,6,7,8,9]

# dictionary that will store our values
samples = {d:[] for d in digits}
file_names = {d:[] for d in digits}

# import files
for filename in os.listdir(data_dir):
    # iterate over digits
    for d in digits:
        if filename.startswith(str(d)+'_'):
            wav = wavfile.read(data_dir+filename)[1]
            if len(wav.shape)<2:
                samples[d].append(wav)
                file_names[d].append(filename)
        
# find maximum of vector length
N = max([len(v) for d in digits for v in samples[d]])
```

## Classification

The dataset has to be split in "training" and testing". The training dataset will be used to determine the average spectrum used in the next classification tasks. The test dataset will be used to evaluate the performance of our method. We create a random split 80/20 split.

```python
# next we split our dataset in train and test
# we will use a 80/20 random split.

# create train/test split
ix = np.arange(100)
random.shuffle(ix)

# select train entries
ix_train = ix[:80]

#select test entries
ix_test = ix[80:]

print(len(samples[2]))
```

Next we compute the average magnitude of the DFT of each digit.

Let $x_{i,d}[n]$ be the set of samples for digit $d$. So $x_{1,1}[n],x_{2,1}[n],\dots,x_{m,1}[n]$ correspond to $m$ samples of digit 1.

We can compute the transform (via the FFT) of each of these signals, given by $$x_{i,d}[n]\leftrightarrow X_{i,d}[k]. $$

We will define the  average spectral magnitude of digit $d$ as 
$$\bar{X}_{d}[k] \triangleq \sum_{i\in \mbox{samples}}\frac{\left|X_{i,d}[k]\right|}{\mbox{number of samples}}.$$

We normalize this quantity to create the average spectral magnitude of digit $d$
$$ 
\tilde{X}_{d}[k]\triangleq\frac{\bar{X}_{d}[k]}{\sqrt{\sum_{k'}\left|\bar{X}_{d}[k']\right|^2}}.$$

## Question 1: Compute $ 
\tilde{X}_{d}[k]$ and store it in the dictionary ``mean_transforms`` below.

```python
# next we compute the average spectrum of each spoken digit in the training set.
# we will consider a window up to 1.5 KHz


# sampling rate is 8kHz
Ts = 1.0/8000

ix_cut = int(np.ceil(1500*Ts*N))

# initialize dictionary for storing transforms
transforms = {d:[] for d in digits}


# initialize dictionary for storing mean transforms
mean_transforms = {d:[] for d in digits}

# compute mean transform for training set
# find the spectrum for each recording
for d in digits:
    for ind in ix_train:
        rec = samples[d][ind]
        rec = np.append(rec, np.zeros(N-len(rec)))
        transforms[d].append(fft(rec))

# find the average spectrum       
for d in digits:
    avg_transform = []
    for freq in range(ix_cut):
        transform_sum = 0
        for rec in transforms[d]:
            transform_sum += np.abs(rec[freq])
        avg_transform.append(transform_sum/len(transforms[d]))
    # normalize the averages
    norm = np.linalg.norm(avg_transform)
    for freq in range(len(avg_transform)):
        mean_transforms[d].append(avg_transform[freq]/norm)
        
```

 ## Question 2: Plot the average spectral maginitude of each digit

```python
for d in digits:
    xf = np.linspace(0.0, 1500, ix_cut)
    yf = mean_transforms[d]
    plt.plot(xf, yf)
    plt.xlabel("Hz")
    plt.grid()
    plt.show()
```

If you selected digits 1 and 2, you can see that there is a very clear distinction between both digits. We create next the mean classifier. For a given signal $x[n]\leftrightarrow X[k]$, the mean_classifier computes $$p\left(X,\tilde{X}_d\right)\triangleq \sum_{k=1}^N \left|X[k]\right|\left|\tilde{X}_d[k]\right|. $$ We refer to this quantity as the `similarity` between $X$ and $\tilde{X}_d$. It quantifies how similar the magnitude of the spectrum of $X$ is with the average spectrum of digit $d$.

## Question 3.a: Write a classifier function that returns the digit for which $x[n]$ has the highest (spectral) similarity.

```python
# classifier function
# receives a vector, computes the product with average digits, and returns the max inner product
# Input: sample x (vector)
def mean_classifier(x):
    similarities = {d:0 for d in digits}
    
    x_extended = np.append(x, np.zeros(N-len(x)))
    x_spec = fft(x_extended)[0:ix_cut]
    x_norm = []
    norm = np.linalg.norm(x_spec)
    for freq in x_spec:
        x_norm.append(freq/norm)
    
    for d in digits:
        sim = 0
        for freq in range(len(x_norm)):
            sim += np.abs(x_norm[freq]) * np.abs(mean_transforms[d][freq])
        similarities[d] = sim
    return max(similarities, key=similarities.get)
```

## Question 3.b: Apply this classifier to our test set and see how it performs.

```python
tot_cor = 0
tot_inc = 0
for d in digits:
    correct = 0
    incorrect = 0
    for ind in ix_test:
        rec = samples[d][ind]
        if(mean_classifier(rec) == d): correct += 1
        else: incorrect += 1
    print(correct)
    print(incorrect)
    tot_cor += correct
    tot_inc += incorrect
    
print("Total Correct: %(a)d, Total Incorrect: %(b)d, Percentage Correct: %(c)f" % {"a": tot_cor, "b": tot_inc, "c": tot_cor/(tot_cor+tot_inc)})
```

## Question 5 (Bonus + 5 points): Can you design a more accurate classifer than the average classifier given above? 

Yes - use the absolute value of the difference between the spectrum of the test audio and the average audio. By replacing the similarity calculation and finding the minimum difference digit, we increase the accuracy by 0.2% to 5%.
