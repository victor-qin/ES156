## Task 1.2 â€“ Plot reconstruction error as a function of noise added

We will use the mean squared error (MSE) as our measurement for reconstruction error. For two $M\times N$ signals $x, x^\prime$, the MSE is defined as: 
$$\frac{1}{N}\frac{1}{M}\sum_{i=0}^{N-1}\sum_{j=0}^{M-1}(x^\prime(i,j) - x(i,j))^2$$

**TODO:**
1. Write a function $\texttt{compute}\text{_}\texttt{MSE}$( ) that computes the mean squared error (MSE) between two images
2. Plot the reconstruction error image as a function of the amount of gaussian noise added 
    - Noises amounts should be $[0,2,4,6,8,10,12,14,16]$
    - Amount of gaussian noise $\equiv$ standard deviation of the noise 
    - Your error should stay below the goal curve


*Example:* the following line of code will create a version of our image that is downsampled and with added gaussian noise of standard deviation of 6:

    im_noisy = utils.noise_image(im_original, 6.0)


```python
def compute_MSE(img, img_new):
    error = 0
    for i in range(N):
        for j in range(N):
            error += (img[i][j] - img_new[i][j])**2
    return error/(N**2)
```

```python
noise_amounts = np.linspace(0,16,8, dtype=np.float64)
errors = []
for i in noise_amounts:
    noisy_im = utils.noise_image(im_original, i, M, N)
    filtered_im = denoise(noisy_im)
    errors.append(compute_MSE(im_original, filtered_im))
goal_curve = [90, 95, 100, 120, 170, 250, 375, 550]
plt.plot(noise_amounts, errors, 'bo--', label='errors')
plt.plot(noise_amounts, goal_curve, 'gs--', label='goal')
plt.legend()
plt.xlabel('Gaussian Noise Added')
plt.ylabel('Reconstruction Error')
plt.show()
```

## Task 2.1 â€“ Get the fourier coefficients of the image
**TODO**: Write a function $\texttt{get}\text{_}\texttt{coeffs}$( ) that gets the fourier coefficients for the image above and returns both the coefficients and their log magnitudes 

*Hint: You may use a function from scipy or numpy to get your fourier coefficients*

```python
def get_coeffs(X):

    # Compute discrete fourier for the image
    cfts = cv2.dft(np.float32(X),flags = cv2.DFT_COMPLEX_OUTPUT)
    
    # Compute the magnitudes of the coefficients
    mgs = 20*np.log(cv2.magnitude(cfts[:,:,0],cfts[:,:,1]))
    
    return cfts, mgs
```

## Task 2.2 â€“ Recover the image using the IFT

**TODO:** Write a function $\texttt{inverse}\text{_}\texttt{ft}$( ) that gets the inverse fourier transform and returns both the coefficients and their magnitudes 

```python
def inverse_ft(cfs):
    # Get the IDFT 
    inv = cv2.idft(cfs)
    
    # Get the magnitudes
    inv_mags = cv2.magnitude(inv[:,:,0],inv[:,:,1])/(M*N)
    
    return inv, inv_mags
```

# Part 3: DCT and JPEG
For image processing applications, it is useful to consider the Discrete Cosine Transform (2D DCT) instead of the 2D DFT due to its superior empirical performance for signal compression and reconstruction tasks. 

We first introduce the two-dimensional discrete cosine $C_{kl}(m, n)$ of frequencies $(k, l)$ defined as:

$$C_{kl, MN}(m,n)= \mbox{cos} \bigg [\frac{k \pi}{2M}(2m+1)\bigg] \bigg[ \frac{l \pi}{2N}(2n+1)\bigg]$$


Then the two-dimensional DCT of a signal $x$ is given by substituting $C_{kl, MN}$ into an expression similar to the two-dimensional DFT, which, after introducing normalization constants, yields:
$$X_C(k,l) := \frac{2}{\sqrt{MN}}\sum_{m=0}^{M-1}\sum_{n=0}^{N-1}x(m,n)c_1c_2 \mbox{cos} \bigg [\frac{k \pi}{2M}(2m+1)\bigg] \bigg[ \frac{l \pi}{2N}(2n+1)\bigg]$$

$$ = \langle x, C_{kl, MN} \rangle $$



where $c_1 = \frac{1}{\sqrt{2}}$ for $k = 0$ and $c_1 = 1$ for $k = 1,...,M-1$ and $c_2 =  \frac{1}{\sqrt{2}}$ for $l = 0$ and $c_2 = 1$ for $l = 1,...,N -1$.

## Task 3.1 â€“ Implement the DCT

**TODO:** Write down a function which takes as input a two-dimensional signal of duration $N^2$ and computes its two-dimensional DCT defined above

```python
# Redefine image dimensions for part 3.1
N = 35
```

```python
def compute_dct(X):
    # Set empty array for dct coefficients
    dct = np.zeros((N, N), dtype=np.float64)

    # Iterate through and compute values for all coefficients
    for k in range(N):
        for l in range(N):

            # Iterate through and compute all values needed for sum of (k,l)th coefficient
            for m in range(N):

                # Set scalar factor values of coefficients
                a1 = 1.0/np.sqrt(2.0) if k == 0 else 1
                a2 = 1.0/np.sqrt(2.0) if l == 0 else 1

                # Compute value for (k,l)th coefficient
                for n in range(N):
                    cm = np.cos((k*np.pi*(2*m+1)/(2*N)))
                    cn = np.cos((l*np.pi*(2*n+1)/(2*N)))
                    dct[k,l] += X[m,n]*a1*a2*cm*cn

    # normalize the transform
    dct = 2*dct/np.sqrt(N**2)
    return dct
```

***Test the results of your DCT function***

We can test whether or not the coefficients returned by our DCT function are correct by comparing it to scipy's DCT function.  Are the output images the same? 

```python
img = cv2.imread("images/dct_sample1.png", 0) 
dct = compute_dct(img)
dct2 = fftpack.dct(fftpack.dct(img.T, type=2, norm='ortho').T, type=2, norm='ortho')


# Plot the image next to the transform and the scipy transform
f = plt.figure(figsize=(12,12))
plt.subplot(131),plt.imshow(img, cmap = 'gray')
plt.xlabel('', fontsize=8), plt.xticks([]), plt.yticks([])
plt.title('Original Image')
plt.subplot(132),plt.imshow(dct, cmap = 'gray')
plt.xlabel('', fontsize=8), plt.xticks([]), plt.yticks([])
plt.title('Results of DCT Function')
plt.subplot(133),plt.imshow(dct2, cmap = 'gray')
plt.xlabel('', fontsize=8), plt.xticks([]), plt.yticks([])
plt.title('Scipys DCT Function')
plt.show()
```

When the signal dimension $N^2$ is very large, it is difficult to represent it well across its entire domain using the same DFT or DCT coeffcients. This is because the computation of the inverse uses the same DCT coeffcients across the entire domain. We address this by partitioning the signal and computing the DCT of each piece so that our DCT coeffcients only needed to locally represent the signal over a small domain.

**TODO:** 
1. Write a function $\texttt{partion}\text{_}\texttt{img}$( ) that takes in a signal (image) of size $N^2$, partitions it into patches of
size $8\times8$ and returns an array of the patches
2. Write a function $\texttt{partioned}\text{_}\texttt{dct}$( ) that takes in an array of patches and a value $K$ and returns an array of the $K^2$ largest DFT coeffcients and their associated frequencies for each patch

*Note:* for our specific case, we are using an image that is $256\times 256$, so we will have $32\times 32$ patches

```python
# Number of patches
prange = N/8
```

```python
def partition_img(img):
    # Set empty array for partioned image
    patches = np.zeros((prange,prange, 8, 8))
    # Split the image into patches of 8x8 
    for i in range(0, N, 8):
        for j in range(0, N, 8):
            patches[i/8,j/8] = img[i:i+8, j:j+8]
    return patches
```

```python
def partitioned_dct(ppatches, K):
    # Set empty array of sizes K^2 per patch
    dct_patches = np.zeros((prange,prange, K, K), dtype=np.float64)
    for i in range(ppatches.shape[0]):
        for j in range(ppatches.shape[1]):
            dct_patches[i,j] = fftpack.dctn(ppatches[i,j], type=2, shape=[K,K], norm='ortho')
    return dct_patches
```

## Task 3.3 â€“ Quantization and JPEG

A rudimentary version of the JPEG compression scheme for images includes partitioning the image into patches, performing the two-dimensional DCT on each patch, and then rounding (or quantizing) the associated DCT coecients. We describe this procedure below:
1. Perform the DCT on each $8\times8$ block.
2. $X_{ij}(k, l) = X_C(x_{ij} )$, where $x_{ij}$ is the $(i, j)$'th block of the image.
3. We then quantize the DCT coeffcients:
$$\hat{X}_{ij}(k,l) = \text{round}\big[\frac{X_{i,j}(k,l)}{Q(k,l)}\big]$$

$Q(k, l)$ is a quantization coeffcient used to control how much the $(k, l)$'th frequency is quantized. Since human vision is not sensitive to these "rounding" errors, this is where the compression takes place. That is, a smaller set of pixel values requires less bits to represent in a computer. The standard JPEG quantization matrix that you should apply to each patch is based upon the way your eye observes luminance, and is given as:
<img src="QuantMatrix.png" style="width:350px;"/> 


**TODO:** Write a function $\texttt{quantize}$( ) that executes the above procedure and returns the quantized coefficients 

```python
# Compute the quantized version of the specified transform (DCT or DFT)
def quantize(dct_patches):

    # Initialize the quantization matrix
    Qmtrx = np.zeros((8,8))
    Qmtrx[0,:] = [16, 11, 10, 16, 24, 40, 51, 61]
    Qmtrx[1,:] = [12, 12, 14, 19, 26, 58, 60, 55]
    Qmtrx[2,:] = [14, 13, 16, 24, 40, 57, 69, 56]
    Qmtrx[3,:] = [14, 17, 22, 29, 51, 87, 80, 62]
    Qmtrx[4,:] = [18, 22, 37, 56, 68, 109, 103, 77]
    Qmtrx[5,:] = [24, 36, 55, 64, 81, 104, 113, 92]
    Qmtrx[6,:] = [49, 64, 78, 87, 103, 121, 120, 101]
    Qmtrx[7,:] = [72, 92, 95, 98, 112, 100, 103, 99]

    # Store dimensions of DCT coefficient matrix patches for iteration
    I = dct_patches.shape[0]
    J = dct_patches.shape[1]
    K = dct_patches.shape[2]
    L = dct_patches.shape[3]

    # Initialize an empty matrix to store the quantized coefficients
    q_dct = np.zeros((I,J,K,L), dtype=np.float64)

    # Compute and store the quantized coefficients
    for i in range(I):
        for j in range(J):
            for k in range(K):
                for l in range(L):
                    q_dct[i,j,k,l] = int(np.round(dct_patches[i,j,k,l])/Qmtrx[k,l])
    return q_dct
```

## Task 3.3 â€“ Unquantization and iDCT:

First, before recovering our image using the inverse DCT, we want to "unquantize" our coefficients by doing the following for all patches $(i,j)$ where $\hat{X}_{ij}$ is the coefficient from the quantized transform and $Q$ is the same standard JPEG quantization matrix from above):
$$X_{ij}(k,l)=\hat{X}_{ij}(k,l) \times Q(k,l)$$




Next, after the un-quantization, we take the inverse DCT which is defined as follows: 
$$\tilde{x}_c(m,n):= \frac{2}{\sqrt{MN}}\sum_{k=0}^{M-1}\sum_{l=0}^{N-1}X_C(k,l)c_1c_2\text{cos}\big[\frac{m\pi (2k+1)}{2M}\big]\big[\frac{n\pi (2l+1)}{2N}\big]$$


Lastly, we must "stitch" our patches back together to get the original image

**TODO:**
1. Write a function $\texttt{unquantize}$( ) which takes in the quantized coefficients for each patch and returns the unquantized coefficients for each patch
2. Write a function $\texttt{inverse}\text{_}\texttt{dct}$( ) which takes in the of unquantized coefficients for each patch, computes the iDCT for each, and returns the recovered patches 
3. Write a function $\texttt{stitch}$( ) that stiches together our recovered patches

```python
def unquantize(qdct_patches):

    # Initialize the quantization matrix
    Qmtrx = np.zeros((8,8))
    Qmtrx[0,:] = [16, 11, 10, 16, 24, 40, 51, 61]
    Qmtrx[1,:] = [12, 12, 14, 19, 26, 58, 60, 55]
    Qmtrx[2,:] = [14, 13, 16, 24, 40, 57, 69, 56]
    Qmtrx[3,:] = [14, 17, 22, 29, 51, 87, 80, 62]
    Qmtrx[4,:] = [18, 22, 37, 56, 68, 109, 103, 77]
    Qmtrx[5,:] = [24, 36, 55, 64, 81, 104, 113, 92]
    Qmtrx[6,:] = [49, 64, 78, 87, 103, 121, 120, 101]
    Qmtrx[7,:] = [72, 92, 95, 98, 112, 100, 103, 99]

    # Store dimensions of DCT coefficient matrix patches for iteration
    I = qdct_patches.shape[0]
    J = qdct_patches.shape[1]
    K = qdct_patches.shape[2]
    L = qdct_patches.shape[3]

    # Initialize an empty matrix to store the quantized coefficients
    iq_dct = np.zeros((I,J,K,L), )

    # Compute and store the quantized coefficients
    for i in range(I):
        for j in range(J):
            for k in range(K):
                for l in range(L):
                    iq_dct[i,j,k,l] = int(qdct_patches[i,j,k,l]*Qmtrx[k,l])
    return iq_dct
```

```python
# Computes iDCTs for patches
def inverse_dct(patches):

    # Set empty array for recovered set of patches image
    i_patches = np.zeros((prange,prange, 8, 8), dtype=np.float64)

    for i in range(prange):
        for j in range(prange):
            i_patches[i,j] = fftpack.idctn(patches[i,j], type=2, shape=[8,8], norm='ortho')

    return i_patches
```

```python
def stitch(patches):
    recovered = np.zeros((N,N), dtype=np.float64)

    for i in range(0, prange):
        for j in range(0, prange):
            recovered[i*8:i*8+8,j*8:j*8+8] = patches[i][j]
    return recovered
```
