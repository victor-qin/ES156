# Part 1: Fourier Image Compressor

How good are Fourier Transform-based methods for image processing? The goal of the first part of this problem is to implement a procedure that takes a $512\times 512$ image as input and compresses it by computing the (two-dimensional) Fourier series and keeping the larger coeffcients. 

You will have to use a two-dimensional Fourier Series of the periodic extension of the image. You don't need to implement this transform itself -- there are many pre-packaged alternatives in Python.

Below is a quick guideline of the steps you will need to produce. Note that the exact implementation of the transform (e.g., the constants used) may depend on the package (e.g., scipy vs numpy) you are importing for computing the transform.


1. **Two-dimensional Fourier series.** You will create a function that takes an image as input and computes the Fourier series of the periodically extended image (see below for details). In particular, two-dimensional Fourier series of a periodic function $x[m,n]$ with period $M$ and $N$, respectively, in the first and the second dimension is given by:
$$
a_{k,l} = \frac{1}{MN} \sum_{m=0}^{M-1} \sum_{n=0}^{N-1} x[m,n] e^{-j\left( \frac{2 \pi}{M}km + \frac{2\pi}{N} ln\right)},
$$  
for $k = 0, \ldots, M-1$ and $l = 0, \ldots, N-1$.

2. **Two-dimensional inverse Fourier series.** A second function will need to produce an image from a subset of the computed of Fourier coefficients above. 

The inverse Fourier series is then given by
$$
x[m,n] = \sum_{k=0}^{M-1} \sum_{l=0}^{N-1} a_{k,l} e^{j\left( \frac{2 \pi}{M}km + \frac{2\pi}{N} ln\right)},
$$  
for $m = 0, \ldots, M-1$ and $n = 0, \ldots, N-1$.

Write a function $\texttt{get}\text{_}\texttt{coeffs}$( ) that gets the fourier coefficients for the image above and returns both the coefficients and their log magnitudes 

*Hint: You may use a function from scipy or numpy to get your fourier coefficients (see cv2.dft)*

```python
def get_coeffs(X):
    # TODO
    temp = np.array(X, dtype='float64')
    cfts = cv2.dft(temp, flags=(cv2.DFT_SCALE + cv2.DFT_COMPLEX_OUTPUT))
#     print(cv2.idft(cfts).shape)
    mgs = 20*np.log10(cv2.magnitude(cfts[:,:,0], cfts[:,:,1]))
    mgs = np.nan_to_num(mgs)
    print(mgs.shape)
    return cfts, mgs
```

# Part 2: DCT and JPEG
For image processing applications, it is useful to consider the Discrete Cosine Transform (2D DCT) instead of the 2D DFT due to its superior empirical performance for signal compression and reconstruction tasks. 

We first introduce the two-dimensional discrete cosine $C_{kl}(m, n)$ of frequencies $(k, l)$ defined as:

$$C_{kl, MN}(m,n)= \mbox{cos} \bigg [\frac{k \pi}{2M}(2m+1)\bigg] \bigg[ \frac{l \pi}{2N}(2n+1)\bigg]$$


Then the two-dimensional DCT of a signal $x$ is given by substituting $C_{kl, MN}$ into an expression similar to the two-dimensional DFT, which, after introducing normalization constants, yields:
$$X_C(k,l) := \frac{2}{\sqrt{MN}}\sum_{m=0}^{M-1}\sum_{n=0}^{N-1}x(m,n)c_1c_2 \mbox{cos} \bigg [\frac{k \pi}{2M}(2m+1)\bigg] \bigg[ \frac{l \pi}{2N}(2n+1)\bigg]$$

$$ = \langle x, C_{kl, MN} \rangle $$



where $c_1 = \frac{1}{\sqrt{2}}$ for $k = 0$ and $c_1 = 1$ for $k = 1,...,M-1$ and $c_2 =  \frac{1}{\sqrt{2}}$ for $l = 0$ and $c_2 = 1$ for $l = 1,...,N -1$.

## Task 2.3  Quantization and JPEG (Optional)

A rudimentary version of the JPEG compression scheme for images includes partitioning the image into patches, performing the two-dimensional DCT on each patch, and then rounding (or quantizing) the associated DCT coecients. We describe this procedure below:
1. Perform the DCT on each $8\times8$ block.
2. $X_{ij}(k, l) = X_C(x_{ij} )$, where $x_{ij}$ is the $(i, j)$'th block of the image.
3. We then quantize the DCT coeffcients:
$$\hat{X}_{ij}(k,l) = \text{round}\big[\frac{X_{i,j}(k,l)}{Q(k,l)}\big]$$

$Q(k, l)$ is a quantization coeffcient used to control how much the $(k, l)$'th frequency is quantized. Since human vision is not sensitive to these "rounding" errors, this is where the compression takes place. That is, a smaller set of pixel values requires less bits to represent in a computer.
